8. Find the number of videos published each month and arrange the months in a decreasing order based

```python
# Count the number of videos published each month
monthly_video_counts = df['published_month'].value_counts().reset_index()
monthly_video_counts.columns = ['Month', 'Video Count']

# Sort the months in decreasing order based on the video count
monthly_video_counts = monthly_video_counts.sort_values(by='Video Count', ascending=False)

# Display the result
print(monthly_video_counts)
```

13. Does the number of views have any effect on how many people disliked the video? Support your
answer with a metric and a plot.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# # Specify the file path to the dataset
# file_path = "youtube_dislike_dataset.csv"

# # Read the dataset into a DataFrame
# df = pd.read_csv(file_path)

# Calculate the correlation between "views" and "Downvotes"
correlation = df['view_count'].corr(df['dislikes'])

# Create a scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
plt.scatter(df['view_count'], df['dislikes'], alpha=0.5)
plt.title('Relationship Between Views and Dislikes')
plt.xlabel('Number of Views')
plt.ylabel('Number of dislikes')
plt.grid(True)

# Add correlation value to the plot
plt.text(df['view_count'].max() * 0.7, df['dislikes'].max() * 0.7,
         f'Correlation = {correlation:.2f}', fontsize=12, color='Blue')

# Show the plot
plt.show()

# Print the correlation value
print(f"Correlation between Views and dislikes: {correlation:.2f}")
print("A positive correlation indicates that as the number of views increases, the number of downvotes also tends to increase.")

```
